```python
import ast
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
intensity_parameter = 5+1
decay_time = 20
pheromone_layers = np.zeros(intensity_parameter)

n_robots = 3
layer = 0
layersEmpty = True
```

```python
# probabilidad random de que pase un robot 
robot = np.random.randint(0,2)
robot = 1
# Compara si pasa un robot por encima del led 
if robot == 1:
    # Recorre el vector de capas buscando la primera capa vacia
    if pheromone_layers[-2] == 0:
        for i in range(0,len(pheromone_layers)-1):
            if pheromone_layers[i] == 0:
                pheromone_layers[i] = decay_time
                layersEmpty = False
                break
    else:
        for i in range(1,len(pheromone_layers)):
            pheromone_layers[i-1] = pheromone_layers[i]
        pheromone_layers[-2]= decay_time


# Disminuye el decay time para cada valor diferente de 0 
for i in range(0,len(pheromone_layers)):
    if pheromone_layers[i] != 0:
        pheromone_layers[i] = pheromone_layers[i] - 1

# Si la primera capa esta vacia y ya ha pasado un robot antes, 
# baja los valores de las demas capas una posici√≥n
if pheromone_layers[0] == 0 and layersEmpty == False:
    for i in range(1,len(pheromone_layers)):
        pheromone_layers[i-1] = pheromone_layers[i]
    print("Se acabo phe") 

# si el vector esta vacio reinicai las variables 
if sum(pheromone_layers) == 0:
    layersEmpty = True

if pheromone_layers[-2] != 0:
    print("pheromo active")

print(pheromone_layers)
```

```python
def obtain_data(csv_file_path,mission_name):
    df = pd.read_csv(csv_file_path)
    df1 =df.set_index('name')

    score1 = df1.loc[f'{mission_name}_100%']['score']
    data1 = ast.literal_eval(score1)

    score2 = df1.loc[f'{mission_name}_80%']['score']
    data2 = ast.literal_eval(score2)

    score3 = df1.loc[f'{mission_name}_60%']['score']
    data3 = ast.literal_eval(score3)

    score4 = df1.loc[f'{mission_name}_40%']['score']
    data4 = ast.literal_eval(score4)

    score5 = df1.loc[f'{mission_name}_20%']['score']
    data5 = ast.literal_eval(score5)
    
    return [data1,data2,data3,data4,data5]
```

```python
def obtain_data2(csv_file_path,mission_name):
    df = pd.read_csv(csv_file_path)
    df1 =df.set_index('name')

    score1 = df1.loc[f'{mission_name}_100%']['score']
    data1 = ast.literal_eval(score1)

    score5 = df1.loc[f'{mission_name}_20%']['score']
    data5 = ast.literal_eval(score5)
    
    return [data1,data5]
```

```python
def plot_data(data, plot_title):
    plt.boxplot(data, notch=True)
    plt.ticklabel_format(style='plain', axis='y', scilimits=(0,0))
    plt.title(plot_title, fontsize=20, family='sans-serif', pad=30)
    plt.xticks(np.array([1,2,3,4,5]),['100%', '80%','60%','40%','20%'], fontsize=15)
    plt.yticks(fontsize=15)
    # plt.savefig("boxplot_aac.png", bbox_inches='tight')
```
